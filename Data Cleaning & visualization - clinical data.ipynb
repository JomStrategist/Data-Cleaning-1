```python
import pandas as pd

data=pd.read_excel('Clinical data analysis.xlsx',skiprows=1)
data
```

```python
data.shape
data.info()
```

```python
data.isna().sum()
```

data contains these much null values in each variable

```python
#filling the date column with most repated date
data['Date']=data['Date'].fillna(data['Date'].mode()[0])
data
```

```python
#filling null values in Height,Weight and Calories with mean.
data.fillna(data[['Height','Weight','Calories']].mean(),inplace=True)

data.head(2)
```

```python
#seperating (splitting) the column Name Doctor in to two variables

data['Name Doctor'.split(' ')]=data['Name Doctor'].str.split(',',expand=True)
data.head(2)
```

```python
#deleting the unwanted column from the data
data.drop('Name Doctor',axis=1,inplace=True)
data
```

```python
data.info()
```

```python
#correcting the wrong entry in the duration column

data.loc[data['Duration']>100,'Duration']=(data.loc[data['Duration']>100,'Duration'].astype(float)/10).astype(int)
data.head(2)
```

```python
#replacing incorrect value in Gender column by correct value.

data.loc[data['Gender']=='Gender','Gender']='Female'

data
```
